```python
def agreeableness_5_to_3(n):
    if n <= 3.17:
        return 0
    if 3.18 <= n <= 3.76:
        return 1
    if n >= 3.77:
        return 2

def neurotizm_5_to_3(n):
    if n <= 2.68:
        return 0
    if 2.69 <= n <= 3.43:
        return 1
    if n >= 3.44:
        return 2

def openness_5_to_3(n):
    if n <= 3.43:
        return 0
    if 3.44 <= n <= 4.08:
        return 1
    if n >= 4.09:
        return 2

def conscientiousness_5_to_3(n):
    if n <= 3.00:
        return 0
    if 3.01 <= n <= 3.67:
        return 1
    if n >= 3.68:
        return 2

def extraversion_5_to_3(n):
    if n <= 3.01:
        return 0
    if 3.02 <= n <= 3.74:
        return 1
    if n >= 3.75:
        return 2

dataset['Экстраверсия_3'] = dataset['Экстраверсия'].apply(lambda n: extraversion_5_to_3(n))
dataset['Доброжелательность_3'] = dataset['Доброжелательность'].apply(lambda n: agreeableness_5_to_3(n))
dataset['Сознательность_3'] = dataset['Сознательность'].apply(lambda n: conscientiousness_5_to_3(n))
dataset['Нейротизм_3'] = dataset['Нейротизм'].apply(lambda n: neurotizm_5_to_3(n))
dataset['Открытость опыту_3'] = dataset['Открытость опыту'].apply(lambda n: openness_5_to_3(n))
```

```python
import re
import emoji


def clean_text(text):
    # приводим текст к нижнему регистру
    text = text.lower()

    # заменяем хештеги на HASHTAG
    text = re.sub(r'(#+[a-zA-Z0-9(_)]{1,})',' HASHTAG ', text)

    # заменяем эмоджи на слова 
    text = emoji.demojize(text, language="en")
    # заменяем эмоджи на EMOJI 
    text = re.sub(r'(:[a-z_]*:)', ' EMOJI ', text)

    # Считаем больше одной повторяющейся скобочки как эмоджи
    text = re.sub(r'[\)\(]+(?=[\)\(])', ' EMOJI ', text)

    # создаем регулярное выражение для замены ссылки на "URL"
    regular_url = r"(http\S+)|(www\S+)|([\w\d]+www\S+)|([\w\d]+http\S+)"
    # заменяем ссылки на "URL"
    text = re.sub(regular_url, " URL ", text)

    # заменяем числа и цифры на ' NUM '
    text = re.sub(r"(\d+\s\d+)|(\d+)", " NUM ", text)

    # создаем регулярное выражение для удаления лишних символов
    regular = r"[\*+\#+\№\"+\+\=+\&\^+\+\>+\<+\\/+\\+\§+\$+\%+\_]"
    # удаляем лишние символы
    text = re.sub(regular, " ", text)

    # удаляем множественные скобочки
    #text = re.sub(r'[\)\.\(]+(?=[\)\.\(])', '', text)

    # заменяем пунктуацию на  ' PUNK '
    text = re.sub(r'[^\w\s]', ' PUNK ', text)

    # удаляем лишние пробелы
    text = re.sub(r"\s+", " ", text)

    # возвращаем очищенные данные
    return text
```

```python
dataset['post_text_preprocessed'] = dataset['post_text'].apply(lambda text: clean_text(text))
dataset
```
