## Логрег предсказывает каждый класс отдельно. Integer от 0 до 4

```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report 
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('../../data/processed/df_tokens_ocean.csv')
dataset.head()
```

```python
dataset['Экстраверсия'] = dataset['Экстраверсия'].apply(lambda x: int(x)-1)
dataset['Доброжелательность'] = dataset['Доброжелательность'].apply(lambda x: int(x)-1)
dataset['Сознательность'] = dataset['Сознательность'].apply(lambda x: int(x)-1)
dataset['Нейротизм'] = dataset['Нейротизм'].apply(lambda x: int(x)-1)
dataset['Открытость опыту	'] = dataset['Открытость опыту'].apply(lambda x: int(x)-1)
dataset.head()
```

# Экстраверсия

```python
PREDICTION_COLUMN = 'Экстраверсия'

X = dataset.tokens
Y = dataset[PREDICTION_COLUMN].apply(lambda x: int(x))
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.33, random_state=42, stratify=Y)

text_transformer = TfidfVectorizer( ngram_range=(1, 2), lowercase=True, max_features=55000)
X_train_text = text_transformer.fit_transform(X_train)
X_test_text = text_transformer.transform(X_test)

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train_text, y_train)
lr_pred = lr.predict(X_test_text)
print(classification_report(y_test, lr_pred))
```

```python
import eli5
eli5.show_weights(estimator=lr,
                  feature_names= list(text_transformer.get_feature_names_out()),
                  top=(50, 5))
```
